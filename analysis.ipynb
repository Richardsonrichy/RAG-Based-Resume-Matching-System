# RAG-Based Resume Matching System

## Overview
This notebook demonstrates a production-ready Resume Matching System using Retrieval-Augmented Generation (RAG).

### Pipeline:
- Resume ingestion (PDF/DOCX)
- Intelligent chunking
- Embedding generation (Sentence Transformers)
- Storage in Chroma Vector DB
- Semantic + Keyword search
- Ranking and scoring

```python
job_description = "Python backend developer with 5 years experience and ML knowledge"

result = match_job(job_description)

print(json.dumps(result, indent=2))
```

```python
queries = [
    "Python developer",
    "Machine learning engineer",
    "Java backend developer",
    "Data scientist",
    "Frontend React developer",
    "DevOps engineer AWS",
    "SQL data analyst"
]

for q in queries:
    print("\n" + "="*60)
    print("QUERY:", q)
    
    result = match_job(q)
    
    if result["top_matches"]:
        top = result["top_matches"][0]
        print("Top Candidate:", top["candidate_name"])
        print("Score:", top["match_score"])
        print("Skills:", top["matched_skills"])
    else:
        print("No matches found")
```

```python
test_query = "Python developer with machine learning"

start = time.time()
result = match_job(test_query)
end = time.time()

print("Latency:", round(end - start, 3), "seconds")
```

```python
query = "Backend Python developer"

result = match_job(query)

print("\nTop Matches:\n")

for i, match in enumerate(result["top_matches"], 1):
    print(f"Rank {i}")
    print("Candidate:", match["candidate_name"])
    print("Score:", match["match_score"])
    print("Skills:", match["matched_skills"])
    print("Reason:", match["reasoning"])
    print("-" * 50)
```
